## Introduction to Linear Regression

-----------------------------

A. Why is Linear Regression important?

Linear regression plays a pivotal role in various fields that touch every aspect of our lives. In life sciences, for example, researchers use linear regression to predict patient outcomes based on medical treatments, helping identify patterns that could save lives. In economic policy, linear regression models are instrumental in forecasting key indicators like GDP growth, inflation, or unemployment, providing policymakers with data-driven insights. Similarly, in business, it aids in revenue forecasting, inventory management, and pricing strategies by analyzing trends over time, leading to more informed decision-making. The ability to model relationships and predict future outcomes makes linear regression a fundamental tool in shaping strategies and solutions across multiple sectors.

B. How does Linear Regression form the foundation of Machine Learning?

Linear regression is often referred to as the “building block” of machine learning, especially in supervised learning. It helps develop algorithms that learn from data to make predictions. In essence, it teaches us the core principle of minimizing errors (through techniques like least squares) to achieve the best model fit. As such, mastering linear regression is the first step in understanding how machines can be trained to recognize patterns and make predictions, forming the backbone of many predictive models used today.

C: How does Linear Regression serve as the foundation for more complex machine learning models?

While linear regression is a fundamental technique, it also serves as the foundation for more complex machine learning models that can capture non-linear relationships and interactions between variables. As machine learning evolves to solve increasingly intricate problems, models such as decision trees, support vector machines, and neural networks build upon the basic principles learned from linear regression. These more advanced models can handle large datasets with high-dimensional features, but they still rely on the underlying concept of minimizing error and optimizing prediction accuracy, a concept first introduced in linear regression. Understanding linear regression is not just an entry point into machine learning but a critical stepping stone toward mastering the full spectrum of predictive modeling techniques used in today’s data-driven world.

To investigate discrepancies in repeat patients, such as different gender or age for the same patient ID, we can check for inconsistencies within repeat patient records. Here’s a step-by-step approach we can take:

- Group by patientid: For each repeat patient, group the dataset by patientid.
- Check for Discrepancies in Key Features: We can identify discrepancies by checking if the Age and gender fields vary for the same patientid.
- Calculate the Percentage of Discrepancies: After identifying these discrepancies, we can calculate the percentage of repeat patients with inconsistent data.

```python
# Step 1: Filter only repeat patients
repeat_patients_data = df_preprocessed[df_preprocessed['patientid'].duplicated(keep=False)]

# Step 2: Group by patientid and check for inconsistencies in gender and age
inconsistent_gender = repeat_patients_data.groupby('patientid')['gender'].nunique() > 1
inconsistent_age = repeat_patients_data.groupby('patientid')['Age'].nunique() > 1

# Step 3: Identify patient IDs with inconsistent data
inconsistent_patients = inconsistent_gender | inconsistent_age

# Step 4: Calculate the percentage of inconsistent patients
total_repeat_patients = repeat_patients_data['patientid'].nunique()
inconsistent_patient_count = inconsistent_patients.sum()

percentage_inconsistent = (inconsistent_patient_count / total_repeat_patients) * 100

print(f"Percentage of repeat patients with inconsistent data: {percentage_inconsistent:.2f}%")
```

A 95.62% inconsistency rate among repeat patients is quite significant. This suggests that the same patientid is being assigned to multiple patients, or the data entry system might be incorrectly logging details like gender and age. This can heavily skew any analysis and conclusions drawn about patient behavior or trends, particularly for repeat patients.

The Chi-Square test will tell you if the overall relationship between these pairs is statistically significant. The residuals will help identify which specific pair deviates the most from expected values, indicating stronger or weaker associations.

##**Linear Regression and OLS Modeling**

**Baseline Model with All Features Retained**

```python
# Step 1: Create a copy of the data for both approaches
X_train_lr_v1 = X_train.copy()  # For Linear Regression model
X_train_ols_v1 = X_train.copy()  # For OLS model

# Step 2: Define a function to calculate regression metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred, X):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    adjusted_r2 = 1 - (1 - r2) * (len(y_true) - 1) / (len(y_true) - X.shape[1] - 1)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R²: {r2}")
    print(f"Adjusted R²: {adjusted_r2}")
    print(f"MAPE: {mape}%")

# Step 3: Fit the Linear Regression model
from sklearn.linear_model import LinearRegression
linear_model_v1 = LinearRegression()
linear_model_v1.fit(X_train_lr_v1, y_train)

# Step 4: Predict on the training set
y_train_pred_v1 = linear_model_v1.predict(X_train_lr_v1)

# Step 5: Evaluate the Linear Regression model
evaluate_model(y_train, y_train_pred_v1, X_train_lr_v1)

# Step 6: Fit the OLS model
import statsmodels.api as sm

# Add constant term for OLS
X_train_ols_v1_const = sm.add_constant(X_train_ols_v1)
ols_model_v1 = sm.OLS(y_train, X_train_ols_v1_const).fit()

# Step 7: Output the OLS summary
print(ols_model_v1.summary())
```

##**OLS (Ordinary Least Squares) Modeling**

The key difference between Linear Regression (sklearn’s implementation) and Ordinary Least Squares (OLS, statsmodels’ implementation) is in the methodology used for fitting the model and the outputs they provide:

1. Library and Functionality
  - sklearn’s Linear Regression is designed for predictive modeling. It focuses on providing predictions, and it supports model evaluation through metrics like RMSE, MAE, and R². It’s typically used in machine learning workflows, especially for handling larger datasets efficiently.
  - statsmodels’ OLS is focused on statistical modeling. It provides a more detailed analysis of the model, such as p-values, confidence intervals, and more robust diagnostic statistics, which are crucial for statistical inferences.

2. Model Diagnostics and Statistics
  - sklearn’s Linear Regression does not give access to detailed statistical information like p-values or confidence intervals. Its goal is mainly predictive accuracy.
  - statsmodels’ OLS provides a full statistical output, including:
    - Coefficients: With p-values, standard errors, and confidence intervals.
    - F-statistic: To test the overall significance of the model.
    - P-values: To assess the significance of each predictor.
    - Durbin-Watson statistic: For checking autocorrelation in residuals.

### **Checking for Multicollinearity**

**Multicollinearity** occurs when **independent variables** in a regression model are **highly correlated** to each other, such that they do not provide **unique** or independent information. A regression coefficient is interpreted as the mean change in the target for each unit change in a feature when all other characteristics are held constant. Changes in one aspect can affect other features when they are correlated. The stronger the relationship, the more difficult it is to change one element without affecting the others. Because the features tend to change concurrently, it becomes difficult for the model to evaluate the link between each variable and the target individually.

**Variation Inflation Factor (VIF)** is one of the most common ways of detecting multicollinearity in data. The Variation Inflation Factor (VIF) is one of the most common methods for detecting multicollinearity in data. The VIF calculates how much the variance of a regression coefficient is inflated due to model multicollinearity.

The VIF can be calculated in two steps,
*   Choose and execute a regression analysis on an independent variable for which you are attempting to calculate VIF. For example, suppose there are three independent variables X1, X2, and X3, and a target variable Y. If we wish to calculate VIF for the variable X1, we use X1 as the target variable and X2 and X3 as independent variables **(X1=b0+b1.X2+b2.X3)**.
* The regression model mentioned above gives us R2 squared. The formula below is used to calculate the VIF using R2.

> ![test_image](https://raw.githubusercontent.com/GL-ACADOPS/RS-IMAGES/main/vif.jpg)

Variance inflation factor (VIFs) tells **"what percentage of the variance is inflated for each coefficient"**. For example, a VIF of 1.7 tells you that the variance of a particular coefficient is 70% bigger than what you would expect if there was no multicollinearity, i.e., if there was no correlation with other predictors.

Usually, features having a VIF score greater than 5 are dropped/treated till all the features have a VIF score of less than 5.
